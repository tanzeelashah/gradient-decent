```python
from sklearn .model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression ,SGDRegressor
```

```python
from sklearn.metrics import r2_score
```

```python
x = 6 * np .random .rand(200,1) -3
y = 0.8 * x**2 + 0.9 * x + 2 + np .random .randn(200,1)
```

```python
plt.plot(x,y,'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
reg =LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(x_train,reg.predict(x_train),color='red')
plt.plot(x,y,'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
ploy = PolynomialFeatures(degree=2)
x_train_trans = ploy.fit_transform(x_train)
x_test_trans = ploy.fit_transform(x_test)
```

```python
print(x_train[0])
print(x_train_trans[0])
```

# include_bias = false toa per zero wala term nahi hogia

```python
reg = LinearRegression()
```

```python
reg.fit(x_train_trans,y_train)
```

```python
y_pred = reg.predict(x_test_trans)
r2_score(y_test,y_pred)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
x_new = np.linspace(-3,3,200).reshape(200,1)
x_new_poly = ploy.transform(x_new)
y_new = reg.predict(x_new_poly)
```

```python

plt.plot(x_new, y_new,"r",linewidth=2, label='predication')
plt.plot(x_train, y_train,'b',label='training piont')
plt.plot(x_test,y_test,'g',label='testing pionts')
plt.plot(x,y,'b')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

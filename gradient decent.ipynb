```python
import numpy as np
```

```python
from sklearn .datasets import make_regression
```

```python
x,y = make_regression (n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
import matplotlib .pyplot as plt
```

```python
plt.scatter(x,y)
```

# oLs apply

```python
from sklearn .linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# take random b 

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

# let apply  gradient decent assuming slope is constant m = 78.35
# and lets assume the starting value for intercept b=0


```python
y_pred = ((78.35*x)+100).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label= 'b=0')
plt.legend()
plt.show()
```

```python
m= 78.35
b=100

loss_slope = -2*np.sum(y-m*x.ravel()-b)
```

```python
loss_slope
```

# let take leaning  rate = 0.1

```python
lr = 0.1
step_size = loss_slope*lr
```

```python
step_size
```

# calcualting new intercept_

```python
b = b-step_size
```

```python
b
```

```python
y_pred1 = ((78.35*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(x,y_pred,color='green',label= 'b=0')
plt.legend()
plt.show()
```

# iteration 2


```python

loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred2 = ((78.35*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='blue',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(x,y_pred,color='green',label= 'b=0')
plt.legend()
plt.show()
```

# 3rd iteration

```python
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred3 = ((78.35*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred3,color='gray',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='blue',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(x,y_pred,color='green',label= 'b=0')
plt.legend()
plt.show()
```

```python
b= -100
m = 78.35
lr = 0.01
epochs = 100
for i in range(epochs):
    loss_slope = -2*np.sum(y-m*x.ravel()-b)
    b = b-(lr*loss_slope)
    y_pred = m*x+b
    plt.plot(x,y_pred)
plt.scatter(x,y)    
```
